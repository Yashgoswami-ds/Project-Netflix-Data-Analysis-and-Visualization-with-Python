```python
'''üü¶ 1. Figure
The overall window or image where everything is drawn.

Think of it like a canvas.
    
fig = plt.figure()
üü© 2. Axes
The part inside the figure where the data is actually plotted.

It contains the X-axis, Y-axis, title, grid, etc.

A figure can have multiple axes (subplots).

ax = fig.add_subplot(1,1,1)
# or easier way
fig, ax = plt.subplots()
üî∑ 3. X and Y Axis
X-axis: Horizontal line (left to right)

Y-axis: Vertical line (bottom to top)

You provide x and y data to plot something:
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]
üî∂ 4. Plot
The actual graph or line you draw using data.
ax.plot(x, y)
üî∏ 5. Marker
Shows the individual data points (e.g., dots, squares, etc.)

Examples:

o = circle

s = square

* = star

ax.plot(x, y, marker='o')

‚ñ™Ô∏è 6. Line Style
The style of the line connecting data points.

Examples:

- = solid line

-- = dashed line

-. = dash-dot line

: = dotted line

ax.plot(x, y, linestyle='--')
üü• 7. Color
The color of the line or marker.

Examples:

'r' = red

'b' = blue

'g' = green

'k' = black

ax.plot(x, y, color='g')
üü® 8. Legend
Shows the label of the lines (especially useful when multiple lines are plotted).
ax.plot(x, y, label='Line 1')
ax.legend()
üè∑ 9. Label
Labels for x-axis and y-axis

ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
üèÅ 10. Title
The title of the graph

ax.set_title('My First Plot')

üîÑ 11. Grid
Adds a grid to the plot (makes it easier to read values).

ax.grid(True)
üìè 12. Axis Limits
Manually set the limits of the x and y axes.

ax.set_xlim(0, 5)
ax.set_ylim(0, 50)

üîÅ 13. Multiple Plots in One Graph
You can draw more than one line in the same plot.

ax.plot(x, y, label='Line 1')
ax.plot(x, [i*2 for i in y], label='Line 2')
ax.legend()
üñºÔ∏è 14. Subplots (Multiple Graphs in One Figure)
Create multiple plots (axes) in a single figure.

fig, (ax1, ax2) = plt.subplots(1, 2)  # 1 row, 2 columns
ax1.plot(x, y)
ax2.plot(x, [i*2 for i in y])
üéØ 15. Annotations
Add text at a specific point on the graph.

ax.annotate('Peak', xy=(3, 25), xytext=(2, 30),
             arrowprops=dict(facecolor='black', shrink=0.05))
üåà 16. Fill Area (Shading)

ax.fill_between(x, y, color='lightblue')

üñç 17. Text
Add custom text anywhere on the plot.

ax.text(2, 20, 'Point Here')
üìä 18. Different Plot Types
Matplotlib can make many plot types, not just line plots:

Bar Chart:

ax.bar(x, y)
Scatter Plot:


ax.scatter(x, y)
Histogram:


ax.hist(y)
Pie Chart:

plt.pie([10, 20, 30], labels=['A', 'B', 'C'])
üß± 19. Tight Layout
Prevents overlapping of titles, labels, etc.


plt.tight_layout()
üíæ 20. Save Plot to File

plt.savefig('myplot.png', dpi=300)


üîπ 1. Method
A method is a function that belongs to an object. In Matplotlib, many actions (like plot(), set_title(), etc.) are methods of objects like Axes or Figure.

Example:

ax.set_title("My Title")  # `set_title` is a method of the `Axes` object
üîπ 2. Parameters
Parameters are values you pass into a method or function to change its behavior.

ax.plot(x, y, color='red', linestyle='--', marker='o')
In this example:

x, y are positional parameters

color, linestyle, and marker are keyword parameters

üîπ 3. Keyword Arguments (kwargs)
A keyword argument is when you pass a value using the name of the parameter.

Example:
ax.plot(x, y, label='Data Line')  # `label` is a keyword argument
Behind the scenes, many plotting functions accept **kwargs, which means ‚Äúany number of optional named arguments‚Äù.

üîπ 4. OO API (Object-Oriented API)
Matplotlib has two main styles:

Pyplot API (easier, like plt.plot())

OO API ‚Äì the recommended way for complex plots, using Figure and Axes objects.

‚úîÔ∏è OO API Example:

import matplotlib.pyplot as plt

fig, ax = plt.subplots()
ax.plot([1, 2, 3], [4, 5, 6])
ax.set_title("Using OO API")
plt.show()
üõë Avoid mixing both styles in complex code!

üîπ 5. dpi (Dots Per Inch)
dpi controls the resolution of the figure. Higher dpi = higher quality image (for print or zoomed viewing).

Example:

plt.savefig("plot.png", dpi=300)
This saves the image at 300 dpi.

üîπ 6. Backend
A backend is what ?. Matplotlib uses to render the plot ‚Äì either to a screen (interactive window), a file, or notebook output.

Some common backends:

Backend	Usage
'TkAgg'	Default for many systems
'Qt5Agg'	Qt-based GUI
'Agg'	Non-GUI (for saving files)
'nbAgg'	Jupyter Notebook

Change backend (not common unless needed):

import matplotlib
matplotlib.use('Agg')  # Use non-interactive backend (for server rendering)
To check current backend:


import matplotlib
print(matplotlib.get_backend())
‚úÖ Summary Table
Term	Meaning
Method	A function inside an object (e.g., ax.plot(), ax.set_title())
Parameters	Values passed to methods/functions
Keyword Args	Named parameters like label="line"
OO API	Object-Oriented approach using fig, ax objects
dpi	Image resolution when saving (dots per inch)
Backend	The rendering engine for plotting
'''
```

```python
import matplotlib.pyplot as plt

# X and Y data for the plot
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

# Create Figure and Axes (Object-Oriented API)
fig, ax = plt.subplots()

# Plot a line with markers, dash-dot style, and green color
ax.plot(x, y, marker='o', linestyle='-.', color='g', label='Line 1')

# Add another line to the same axes
ax.plot(x, [i * 2 for i in y], marker='s', linestyle='-', color='b', label='Line 2')

# Add legend to distinguish lines
ax.legend()

# Add X and Y axis labels
ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")

# Add a title to the plot
ax.set_title("My First Plot")

# Show gridlines
ax.grid(True)

# Set limits for X and Y axes
ax.set_xlim(0, 5)
ax.set_ylim(0, 60)

# Annotate a point with an arrow
ax.annotate('Peak', xy=(3, 25), xytext=(2, 45),
            arrowprops=dict(facecolor='black', shrink=0.05))

# Fill the area under the curve
ax.fill_between(x, y, color='lightblue', alpha=0.3)

# Add custom text inside the plot
ax.text(2, 20, 'Point Here')

# Adjust layout to prevent overlaps
plt.tight_layout()

# Save figure to file with 300 DPI (high resolution)
plt.savefig('basic_line_plot.png', dpi=300)

# Show the plot window
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import autumn

'''Bar chart
Scatter plot
Histogram
Pie chart'''

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

'''
+-----------+-----------+-----------+
| axs[0, 0] | axs[0, 1] | axs[0, 2] |
+-----------+-----------+-----------+
| axs[1, 0] | axs[1, 1] | axs[1, 2] |
+-----------+-----------+-----------+
| axs[2, 0] | axs[2, 1] | axs[2, 2] |
+-----------+-----------+-----------+
'''

# Create a 3x3 grid of subplots
fig, axs = plt.subplots(3, 3, figsize=(12, 10))

# Bar chart
axs[0, 0].bar(x, y)
axs[0, 0].set_title("Bar chart")

# Scatter chart
axs[0, 1].scatter(x, y, color="red")
axs[0, 1].set_title("Scatter chart")

# Pie chart inside axs[0,2]
axs[0, 2].pie([10, 20, 30], labels=['A', 'B', 'C'], autopct='%1.1f%%')
axs[0, 2].set_title("Pie chart")

# Histogram
axs[1, 0].hist(y, bins=10, color="orange")
axs[1, 0].set_title("Histogram chart")

# Line chart
axs[1, 1].plot(x, y, marker='o', linestyle='--', color="green")
axs[1, 1].set_title("Line chart")

# Box plot
axs[2, 0].boxplot(y)
axs[2, 0].set_title("Box Plot")

# Area chart
axs[2, 1].fill_between(x, y, color="skyblue", alpha=0.5)
axs[2, 1].plot(x, y, color="blue")
axs[2, 1].set_title("Area Chart")

# Hide unused subplots (optional) . currently not in used 
axs[1, 2].axis('off')
axs[2, 2].axis('off')

# ‚úÖ Adjust layout BEFORE showing
plt.tight_layout()

# ‚úÖ Show the complete figure
plt.show()
```

```python
  PYPLOT FUNCTION

grid , show . xlim ,ylim , xticks ,yticks, legend, title , plot ,xlabel,ylabel



üìä Common Pyplot Functions
1. plot()

Used to plot data points on a 2D graph.

Syntax:

plt.plot(x, y, label='Line Name')

2. xlabel() & ylabel()

Add labels to the X-axis and Y-axis.

Syntax:

plt.xlabel("X-Axis Label")
plt.ylabel("Y-Axis Label")

3. title()

Adds a title to the plot.

Syntax:

plt.title("Title of the Graph")

4. legend()

Displays the legend for labeled lines or plots.

Required if you use label inside plot().

Syntax:

plt.legend()

5. grid()

Adds a grid to the background of the plot for better readability.

Syntax:

plt.grid(True)

6. xlim() & ylim()

Set limits for the X and Y axes.

Syntax:

plt.xlim(min, max)
plt.ylim(min, max)

7. xticks() & yticks()

Set the tick marks on the X and Y axes.

Syntax:

plt.xticks([list_of_ticks])
plt.yticks([list_of_ticks])

8. show()

Displays the final plotted graph.

Always use at the end.

Syntax:

plt.show()
```

```python
import pandas as pd 
import  matplotlib.pyplot as plt

# load csv
df = pd.read_csv('Netflix_MetaData.csv')

df=df.dropna(subset=['type' ,'date_added','rating','country','duration'])
type_counts=df['type'].value_counts()
plt.figure(figsize=(6,4))
plt.bar(type_counts.index,type_counts.values,color=['skyblue','orange'])
plt.xlabel('Type')
plt.ylabel('count')
plt.tight_layout()
plt.savefig('Movie_VS_Series.png')



raringCount= df['rating'].value_counts()
plt.figure(figsize=(14,12))
plt.pie(raringCount,labels=raringCount.index,autopct='%1.1f%%',startangle=90)
plt.title('percentage of content rating')
plt.tight_layout()
plt.savefig('rating.png')



# Filter only Movies
movie_df = df[df['type'] == 'Movie'].copy()

# Remove 'min' and convert to int
movie_df['duration_int'] = movie_df['duration'].str.replace('min', '').str.strip()
movie_df = movie_df.dropna(subset=['duration_int'])  # Drop missing durations if any
movie_df['duration_int'] = movie_df['duration_int'].astype(int)

# Plot histogram
plt.figure(figsize=(6,4))
plt.hist(movie_df['duration_int'], bins=40, color='purple', edgecolor='black')
plt.title('Distribution of Movie Durations')
plt.xlabel('Duration in Minutes')
plt.ylabel('Number of Movies')
plt.tight_layout()
plt.savefig('Movie_Dur.png')
plt.show()

# scatter
releaseCount= df['release_year'].value_counts().sort_index()
plt.figure(figsize=(10,6))
plt.scatter(releaseCount.index, releaseCount.values,color='red')
plt.title('Release year vs no of shows')
plt.xlabel('release year')
plt.ylabel('Number of Shows')
plt.tight_layout()
plt.savefig('ReYear_VS_Show.png')
plt.show()

# top 10 country have provided high shows(horizontal bar)
CountryCount=df['country'].value_counts().head(10)
plt.figure(figsize=(8,6))
plt.barh(CountryCount.index,CountryCount.values,color='teal')
plt.title('Top 10   Country by no of shoes')
plt.xlabel('no of shows')
plt.ylabel('Country')
plt.tight_layout()
plt.savefig('Top10country.png')
plt.show()

# content by year used group by year
content_by_year=df.groupby(['release_year','type']).size().unstack().fillna(0)

fig, ax= plt.subplots(1,2,figsize=(12,5))
# 1st subplot:movies.
ax[0].plot(content_by_year.index,content_by_year['Movie'],color='blue')
ax[0].set_xlabel('year')
ax[0].set_ylabel('Number of movies')

# 2nd subplot:tvshow.
ax[0].plot(content_by_year.index,content_by_year['TV Show'],color='orange')
ax[0].set_xlabel('year')
ax[0].set_ylabel('Number of movies')
fig.suptitle('com of movie and tv show over the year')
plt.tight_layout()
plt.savefig('moviesTvshows_comp.png')
plt.show()

```
